```python
data = pd.read_csv('/home/cbmr/kzd307/gitte/hippocampus/data/zhong_cluster8.csv')
```

```python
data['V1'].tolist()
```

```python
data.rename(columns={'Unnamed: 0':'bla'}, inplace=True )
data = data.set_index('bla')
data.index.name = None
```

```python
# Convert genes from mouse to human
r = requests.post(
    url='https://biit.cs.ut.ee/gprofiler/api/orth/orth/',
    json={
        'organism':'hsapiens',
        'target':'ENSG',
        'query':data['V1'].tolist(),
    }
    )
```

```python
human_id = pd.DataFrame(r.json()['result'])
```
